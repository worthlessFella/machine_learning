```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data
x_input = np.zeros((20, 3))
x_input[:, 0] = 1
x_input[:10, 1] = 1
x_input[10:, 2] = 1
# label first five points as 0 and rest as 1
label = np.array([0]*10 + [1]*10)

print("input:", x_input)
print("label:", label)
```

```python
losses = [] # keep track of squared error cost
weight = np.random.randn(3) / np.sqrt(3) # randomly initialize w
learning_rate = 0.001
for i in range(1000):
  # forward model
  y_hat = x_input.dot(weight)
  delta = y_hat - label
  weight = weight - learning_rate * x_input.T.dot(delta)

  mse_loss = delta.dot(delta) / 10
  losses.append(mse_loss)
```

```python
# plot the losses
plt.plot(losses)
plt.show()

print("final weight:", weight)

# plot
plt.plot(y_hat, label='preds')
plt.plot(label, label='label')
plt.legend()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 30
X = np.linspace(0,10,N)
Y = 0.5*X + np.random.randn(N)

print('X, Y shape', X.shape, Y.shape)

# plot the data
plt.scatter(X, Y)
plt.show()
```

```python
epochs = 10
# initialise weight
W = np.random.randn(1) # M=1 as number of feature is 1
print('W X.shape', W.shape), X.shape
learning_rate = 0.01
losses = []


# add bias term
B = np.ones((1))

X = X.reshape(-1, 1)
W = W.reshape(-1, 1)
B = B.reshape(-1, 1)
Y = Y.reshape(-1, 1)
print('X W B Y', X.shape, W.shape, B.shape, Y.shape)

for idx in range(epochs):
  Y_hat = X.dot(W) + B
  error = Y_hat - Y

  loss = (1 / (2 * N)) * np.sum(error ** 2)

  gradient = (1 / N) * (X.T.dot(error))

  W = W - learning_rate * gradient
  B = B - learning_rate * gradient

  losses.append(loss)
```

```python
# print(W[1].shape)
Y_pred = X.dot(W) + B
print('X.shape Y_pred', X.shape, Y_pred.shape)

plt.scatter(X, Y_pred, color='r')
plt.scatter(X, Y, color='b')
```

```python
plt.plot(losses, range(epochs))
```
